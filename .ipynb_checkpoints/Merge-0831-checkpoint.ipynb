```python
#dependencies and option setup
import pandas as pd
from sqlalchemy import create_engine
```

```python
#import csv
happiness = pd.read_csv('2015 Happiness Data.csv')
wdi = pd.read_csv('wdi_2015.csv')
```

```python
#determine approximate number of indicators
x=list(wdi['IndicatorName'].unique())
indicators= [x[0]]
for i in range(1, len(x)-1):
    if x[i+1].split()[0:2] != x[i].split()[0:2]:
        indicators.append(x[i+1])
print(f'approximation of number of indicators: {len(indicators)}')
indicators
```

```python
#pivot table created on Country Name and index reset to create primary key
indicator_table = pd.pivot_table(wdi, values='Value', index='CountryName', columns='IndicatorCode')
indicator_table.reset_index(inplace=True)
indicator_table
```

```python
#merge wdi with happiness df
merge = pd.merge(happiness, indicator_table, left_on = 'Country', right_on = 'CountryName', how='inner')
merge.drop(columns='CountryName', inplace=True)
merge
```

```python
#display indicators and their indicator codes
indicate=wdi.iloc[:,2:4]
indicate.drop_duplicates(inplace=True)
#reset index
reset_index_df = indicate.reset_index()
indicator_df = reset_index_df.drop(['index'], axis=1)

#display max rows
pd.set_option('display.max_rows', None)

#display df
indicator_df
```

```python
#reset option
pd.set_option('display.max_rows', 15)

#world happiness dataframe merged with development indicators
column_list1 = ['Country', 'Region', 'Happiness Rank', 'Happiness Score',
       'Standard Error', 'Economy (GDP per Capita)', 'Family',
       'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)',
       'Generosity', 'Dystopia Residual']
df1=merge.loc[:, column_list1]
#list of indicators the group collectively determines as interesting/relevant
column_list2 = ['FB.BNK.CAPA.ZS', 'IC.ELC.OUTG', 'IC.FRM.OUTG.ZS', 'SH.STA.ACSN', 'SH.H2O.SAFE.ZS', 'IC.TAX.TOTL.CP.ZS', 'SI.POV.NAHC', 'IC.FRM.CRIM.ZS', 'IC.FRM.FEMM.ZS', 'IC.FRM.CORR.ZS', 'IC.FRM.OUTG.ZS']
df2=merge.loc[:, column_list2]
final = pd.concat([df1, df2], axis=1)
final
```

```python
#notice numerous null values in the above dataframe - for loop used to further inspect data
for i in range(12, len(final.columns)):
    name = final.iloc[:,i].name
    null_sum = final.iloc[:,i].isna().sum()
    print(f"{name} has {null_sum} null values")
```

```python
#appends list where null values are less than 10
code_list=[]
for i in range(13,len(merge.columns)):
    name=merge.iloc[:,i].name
    null_sum=merge.iloc[:,i].isna().sum()
    if null_sum <10:
        code_list.append(name)
        print(f"{name} has {null_sum} null values")
```

```python
#merges happiness df with columns that contain less than 10 null values
df1=merge.loc[:,'Country':'Dystopia Residual']
df2=merge.loc[:,code_list]
final_subset = pd.concat([df1, df2], axis=1)
final_subset
```

```python
#display max rows
pd.set_option('display.max_rows', None)
#determine indicator code and respective indicator name for those with less than 10 null values
reset_index_df = indicator_df[indicator_df['IndicatorCode'].isin(code_list)].reset_index()
code_list_df = reset_index_df.drop(['index'], axis=1)
code_list_df
```
