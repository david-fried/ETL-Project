```python
import pandas as pd
pd.options.display.max_columns = None
pd.options.display.max_rows = None
happiness = pd.read_csv('2015 Happiness Data.csv')
wdi = pd.read_csv('wdi_2015.csv')
```

```python
x=list(wdi['IndicatorName'].unique())
indicators= [x[0]]
for i in range(1, len(x)-1):
    if x[i+1].split()[0:2] != x[i].split()[0:2]:
        indicators.append(x[i+1])
print(f'approximation of number of indicators: {len(indicators)}')
indicators
```

```python
table = pd.pivot_table(wdi, values='Value', index='CountryName', columns='IndicatorCode')
table.reset_index(inplace=True)
table
```

```python
merge = pd.merge(happiness, table, left_on = 'Country', right_on = 'CountryName', how='inner')
#merge.drop(columns='CountryName', inplace=True)
merge
```

```python
#'FB.BNK.CAPA.ZS', 'IC.ELC.OUTG', 'IC.FRM.OUTG.ZS', 'SH.STA.ACSN', 'SH.H2O.SAFE.ZS', 'IC.TAX.TOTL.CP.ZS',
#'SI.POV.NAHC', 'IC.FRM.CRIM.ZS', 'IC.FRM.FEMM.ZS', 'IC.FRM.CORR.ZS', 'IC.FRM.OUTG.ZS'
merge['FB.BNK.CAPA.ZS'].isna().sum()
```

```python
merge['IC.FRM.OUTG.ZS'].isna().sum()
```

```python
code_list=[]
for i in range(13,len(merge.columns)):
    name=merge.iloc[:,i].name
    null_sum=merge.iloc[:,i].isna().sum()
    if null_sum <10:
        code_list.append(name)
        print(f"{name} has {null_sum} null values")
```

```python
code_list
```

```python
df1=merge.loc[:,'Country':'Dystopia Residual']
df2=merge.loc[:,code_list]
final_subset = pd.concat([df1, df2], axis=1)
final_subset.to_csv("Resources/data.csv")
final_subset
```
