```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress

Data = "2015 Happiness Data.csv"
```

```python
HappyData = pd.read_csv(Data)
HappyData.head()
```

```python
# Analysis

(slope, intercept, rvalue, pvalue, stderr) = linregress(HappyData["Happiness Score"], HappyData["Freedom"])
regressValues = HappyData["Happiness Score"]* slope + intercept
lineEq = f"y = {round(slope, 2)} x + {round(intercept, 2)}"

# CoefCor = x.corr(y)
print(f"Coef. of Corr. Happiness Score vs Freedom = {round(rvalue,2)}")

plt.figure(figsize = (8,7))

plt.scatter(HappyData["Happiness Score"], HappyData["Freedom"])
plt.plot(HappyData["Happiness Score"], regressValues, color = "green")
plt.annotate(lineEq,(5.5,0.1), color = "red", fontsize=16)
plt.xlabel("Happiness Score")
plt.ylabel("Freedom")
plt.title("Happiness vs Freedom")

plt.show()
```

```python
(slope, intercept, rvalue, pvalue, stderr) = linregress(HappyData["Happiness Score"], HappyData["Economy (GDP per Capita)"])
regressValues = HappyData["Happiness Score"]* slope + intercept
lineEq = f"y = {round(slope, 2)} x + {round(intercept, 2)}"

# CoefCor = x.corr(y)
print(f"Coef. of Corr. Happiness Score vs Economy (GDP) = {round(rvalue,2)}")
print(f"There is a strong relationship between happiness and the performance of the economy")
plt.figure(figsize = (8,7))

plt.scatter(HappyData["Happiness Score"], HappyData["Economy (GDP per Capita)"])
plt.plot(HappyData["Happiness Score"], regressValues, color = "green")
plt.annotate(lineEq,(5.5,0.1), color = "red", fontsize=16)
plt.xlabel("Happiness Score")
plt.ylabel("Economy (GDP)")
plt.title("Happiness vs Economy (GDP)")

plt.show()
```
